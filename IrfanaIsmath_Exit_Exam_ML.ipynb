## Handling missing values

```python
#checking overall percent of missing values in the data
total_cells = np.product(data.shape)
total_missing = missing_values.sum()
percent_missing_vals = ((total_missing/total_cells)*100).round(2)
print('Overall percentage of missing values: ', percent_missing_vals)
```

```python
# Outlier handling in ball_controlling_skills 
Q1_ball_controlling_skills=data['ball_controlling_skills'].quantile(0.25)
Q3_ball_controlling_skills=data['ball_controlling_skills'].quantile(0.75)
IQR_ball_controlling_skills=Q3_ball_controlling_skills-Q1_ball_controlling_skills
upper_ball_controlling_skills=Q3_ball_controlling_skills+(1.5*IQR_ball_controlling_skills)
lower_ball_controlling_skills=Q1_ball_controlling_skills-(1.5*IQR_ball_controlling_skills)
data['ball_controlling_skills']=np.where(data['ball_controlling_skills']>upper_ball_controlling_skills,upper_ball_controlling_skills,np.where(data['ball_controlling_skills']<lower_ball_controlling_skills,lower_ball_controlling_skills,data['ball_controlling_skills']))
```

```python
Q1_body_reflexes = data['body_reflexes'].quantile(0.25)
Q3_body_reflexes = data['body_reflexes'].quantile(0.75)
IQR_body_reflexes = Q3_body_reflexes - Q1_body_reflexes
upper_body_reflexes = Q3_body_reflexes + (1.5 * IQR_body_reflexes)
lower_body_reflexes = Q1_body_reflexes - (1.5 * IQR_body_reflexes)

data['body_reflexes']=np.where(data['body_reflexes']>upper_body_reflexes,upper_body_reflexes,np.where(data['body_reflexes']<lower_body_reflexes,lower_body_reflexes,data['body_reflexes']))
```

```python


Q1_body_balance = data['body_balance'].quantile(0.25)
Q3_body_balance = data['body_balance'].quantile(0.75)
IQR_body_balance = Q3_body_balance - Q1_body_balance
upper_body_balance = Q3_body_balance + (1.5 * IQR_body_balance)
lower_body_balance = Q1_body_balance - (1.5 * IQR_body_balance)

data['body_balance'] = np.where(data['body_balance'] > upper_body_balance,
                                 upper_body_balance,
                                 np.where(data['body_balance'] < lower_body_balance,
                                          lower_body_balance,
                                          data['body_balance']))
```

```python

Q1_jumping_skills = data['jumping_skills'].quantile(0.25)
Q3_jumping_skills = data['jumping_skills'].quantile(0.75)
IQR_jumping_skills = Q3_jumping_skills - Q1_jumping_skills
upper_jumping_skills = Q3_jumping_skills + (1.5 * IQR_jumping_skills)
lower_jumping_skills = Q1_jumping_skills - (1.5 * IQR_jumping_skills)

data['jumping_skills'] = np.where(data['jumping_skills'] > upper_jumping_skills,
                                  upper_jumping_skills,
                                  np.where(data['jumping_skills'] < lower_jumping_skills,
                                           lower_jumping_skills,
                                           data['jumping_skills']))
```

```python


# List of columns to handle outliers
columns_to_handle = ['age','penalties_conversion_rate', 'mental_strength',
                     'goalkeeping_skills', 'passing_skills',
                     'dribbling_skills', 'body_strength_stamina',
                     'max_running_speed', 'behaviour_rating', 'matches_played',
                     'fitness_rating', 'years_of_experience',
                     'no_of_disqualifications']

# Loop through each column and apply the outlier handling code
for column in columns_to_handle:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + (1.5 * IQR)
    lower_bound = Q1 - (1.5 * IQR)
    
    data[column] = np.where(data[column] > upper_bound,
                            upper_bound,
                            np.where(data[column] < lower_bound,
                                     lower_bound,
                                     data[column]))
```
